```python
from googletrans import Translator
import pandas as pd
import time

translator = Translator()

def translate_caption(caption):
    if pd.isna(caption) or caption.strip() == '':
        return caption, False

    try:
        detected_language = translator.detect(caption).lang
        if detected_language != 'en':
            time.sleep(0.1)  # Delay to prevent rapid requests
            translated_caption = translator.translate(caption, dest='en').text
            return translated_caption, True
        else:
            return caption, False
    except Exception as e:
        print(f"Error in translation: {e}")
        return caption, False

# Apply the translation to the DataFrame
data['Translated_Caption'], data['Is_Translated'] = zip(*data['caption'].apply(translate_caption))

# Save the updated DataFrame
data.to_csv('/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/Metadata/test.csv', index=False)
```

```python
from googletrans import Translator
import pandas as pd

translator = Translator()

def translate_caption(caption):
    # Check if the caption is empty or NA
    if pd.isna(caption) or caption == '':
        return caption, False  # Return the original caption and False for 'Is Translated'

    # Detect the language of the caption
    detected_language = translator.detect(caption).lang

    # Translate if the caption is not in English
    if detected_language != 'en':
        try:
            translated_caption = translator.translate(caption, dest='en').text
            return translated_caption, True
        except Exception as e:
            print(f"Error in translation: {e}")
            return caption, False
    else:
        return caption, False

# Test the function with a sample caption
sample_caption = "hola"
translated_caption, is_translated = translate_caption(sample_caption)
print("Original:", sample_caption)
print("Translated:", translated_caption)
print("Is Translated:", is_translated)
```

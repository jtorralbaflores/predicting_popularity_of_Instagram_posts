```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/Metadata/all_metadata_3.csv', lineterminator='\n')
```

```python
columns_to_drop = ['media_preview', 'Profile_pic', 'Phone', 'Email', 'caption_is_edited', 'owner_id', 'shortcode']
data = data.drop(columns=columns_to_drop)
```

```python
print(data.shape)
```

```python
# Merge tranlsation data

# Load translatin data
translation_data = pd.read_csv('/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/Metadata/test.csv', lineterminator='\n')
```

```python
selected_columns = ['Is_English', 'Detected_Language', 'post_id']
translation_data = translation_data[selected_columns]
```

```python
print(translation_data)
```

```python
merged_data = data.merge(translation_data, on='post_id', how='left')
```

```python
print(merged_data.head())
```

```python
# Assuming your merged DataFrame is named 'merged_data'

na_count_is_english = merged_data['Is_English'].isna().sum()
na_count_detected_language = merged_data['Detected_Language'].isna().sum()

print("Number of NAs in 'Is_English':", na_count_is_english)
print("Number of NAs in 'Detected_Language':", na_count_detected_language)
```

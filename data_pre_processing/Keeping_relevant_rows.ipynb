```python
import numpy as np
import pandas as pd

filenames = []
for i in range(1, 15):  # Adjust the range based on your files
    batch_filenames = np.load(f'/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/images/filenames_batch{i}.npy')
    filenames.extend(batch_filenames)
```

```python
# Convert the list of filenames to a DataFrame
filenames_df = pd.DataFrame(filenames, columns=['Image files'])

print(filenames_df.head)
print(filenames_df.shape)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/Metadata/all_metadata_4.csv', lineterminator='\n')
```

```python
print(data.shape)
print(data.head)
```

```python
data['Image_Count'] = data.groupby('post_id')['Image files'].transform('count')
```

```python
# Filter the metadata DataFrame to keep only rows with matching filenames
filtered_metadata = data[data['Image files'].isin(filenames_df['Image files'])]
```

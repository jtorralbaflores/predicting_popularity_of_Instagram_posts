```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Loading the data
data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/MSc thesis/processed_data/metadata/fully_concatenated_data.csv", lineterminator='\n')
```

Creating like proportion visualizations

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['like_count'], bins=50, kde=True)
plt.title('Distribution of Like Count')
plt.xlabel('Like Count')
plt.ylabel('Frequency')
plt.show()
```

```python
top_1_percentile = data['like_count'].quantile(0.99)
top_5_percentile = data['like_count'].quantile(0.95)

print(f"Top 1% of likes start at: {top_1_percentile}")
print(f"Top 5% of likes start at: {top_5_percentile}")
```

```python
# For dropping top 1% data
trimmed_data = data[data['like_count'] <= top_1_percentile]

plt.figure(figsize=(10, 6))
sns.histplot(trimmed_data['like_count'], bins=50, kde=True)
plt.title('Distribution of Like Count (top 1% of data cutoff)')
plt.xlabel('Like Count')
plt.ylabel('Frequency')
plt.show()
```

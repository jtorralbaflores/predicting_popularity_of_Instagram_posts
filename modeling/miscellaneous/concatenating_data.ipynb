Now we match the image files from the metadata with the image files from the extracted features.

Let's match the image features witht the metadata

```python
# Load the metadata
metadata = pd.read_csv('/content/drive/MyDrive/MSc Thesis Data Science/influencer_brands/Data/Metadata/all_metadata_5.csv', lineterminator='\n')
```

```python
print(metadata['Sponsorship label'].unique())
```

```python
# Load the image features and filenames
image_features = np.load('/content/drive/MyDrive/Colab Notebooks/MSc thesis/processed_data/images/concatenated_data/image_features_concatenated.npy')
filenames = np.load('/content/drive/MyDrive/Colab Notebooks/MSc thesis/processed_data/images/concatenated_data/image_filenames_concatenated.npy')

# Convert the image features numpy array into a DataFrame
image_features_df = pd.DataFrame(image_features)

# Create a DataFrame from filenames
filenames_df = pd.DataFrame(filenames, columns=['Image files'])

# Renaming columns with a prefix 'image'
image_feature_columns = {i: f'image_{i}' for i in image_features_df.columns}
image_features_df.rename(columns=image_feature_columns, inplace=True)

# Combine filenames and features into a single DataFrame
image_data = pd.concat([filenames_df, image_features_df], axis=1)

# Merge this with your metadata on filenames
metadata = metadata.merge(image_data, on='Image files', how='left')

# Display the first few rows of the merged DataFrame
print(metadata.head())
```

```python
text_data = pd.concat([text_filenames_df,text_features_df], axis = 1)
```
